```python
# importing libraries
import seaborn as sns, pandas as pd, matplotlib.pyplot as plt, numpy as np

sns.set(color_codes=True)
```

```python
tips_df = pd.read_csv("tips.csv")
tips_df.head(2)
```

```python
# plot the relationship between two variables in a df
sns.regplot(x="total_bill", y="tip", data=tips_df)
plt.show()
```

```python
# plot the two variable defined as arrays using different color
np.random.seed(8)
mean=[4,6]
cov=[(1.5,0.7),(0.7,1)]
x_value, y_value = np.random.multivariate_normal(mean,cov,80).T
sns.regplot(x=x_value, y=y_value, color='g')
plt.show()
```

```python
# plot two variable using pandas , with different colors
x_value, y_value = pd.Series(x_value,name="x_var"), pd.Series(y_value, name="y_var")
sns.regplot(x=x_value, y=y_value, marker="x")
plt.show()
```

```python
# use confidence interval with 50%
sns.regplot(x_value,y_value,ci=50)
plt.show()
```

```python
# plot with a discrete x variable and add some jitter
sns.regplot(x="size", y="total_bill", data=tips_df,x_jitter=.1)
plt.show()
```

```python
# plot with x variable showing means
sns.regplot(x="size", y="total_bill", data=tips_df, x_estimator=np.mean)
plt.show()
```

```python
# plot with a continous variable divided into discrete bins
sns.regplot(x=x_value, y=y_value, x_bins=4)
plt.show()
```

```python
#fit a higher order polynomial regression and truncate the model prediction
anscombe_df = sns.load_dataset("anscombe")
```

```python
anscombe_df.head(2)
```

```python
sns.regplot(x="x",y="y",data=anscombe_df.loc[anscombe_df.dataset=="II"], order=2,ci=None, truncate=True)
plt.show()
```

```python
sns.regplot(x="x", y="y", data = anscombe_df.loc[anscombe_df.dataset == "III"], robust=True)
plt.show()
```

```python
tips_df["big_tip"] = (tips_df.tip/ tips_df.total_bill) > 0.175
sns.regplot(x="total_bill", y="big_tip", data=tips_df, logistic=True, y_jitter=0.03)
plt.show()
```

```python
sns.regplot(x="size", y="total_bill", data=tips_df, x_estimator=np.mean, logx=True, truncate=True)
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip",data=tips_df, marker="D",\
           scatter_kws={'color':"blue"},\
           line_kws={"color":"red","linewidth":4.5})
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips_df, marker='o',\
           scatter_kws={'color':'gray'},\
           line_kws={'color':'black', 'linewidth':2.5})
plt.show()
```
