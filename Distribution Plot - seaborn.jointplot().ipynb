# Distribution Plot

## seaborn.jointplot()

> seaborn.jointplot()

Draw a plot of two variable with bivariate and univariate graphs

> Paremeters
  1. x, y :- strings or vectors
  2. data :- dataframe, optional
  3. kind :- scatter, reg, resid, kde, hex, optional
  4. stat_function :- callable or none
  5. color :- matplotlib color, optional
  6. height :- nummberic , optional
  7. ratio :- numeric , optional
  8. space :- numeric, optional
  9. dropna :- bool , optional
  10. xlim, ylim :- two tuples , optional
  11. joint_kws, marginal_kws, annot_kws :- dicts, optional
  12. kwargs:- key, value pairings

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
```

```python
# draw a scatterplot with marginal histograms
sns.jointplot(x='total_bill', y='tip', data= tips_df)
plt.show()
```

```python
# draw a scatterplot with marginal histograms add kde fits
sns.jointplot(x='total_bill', y='tip', data= tips_df, kind='kde')
plt.show()
```

```python
# draw a scatterplot with marginal histograms add regression
sns.jointplot(x='total_bill', y='tip', data= tips_df, kind='reg')
plt.show()
```

```python
# replace scatterplot with a joint histograms with hexagonal bins
sns.jointplot(x='total_bill', y='tip', data= tips_df, kind='hex')
plt.show()
```

```python
iris_df = sns.load_dataset("iris")
```

```python
# draw a scatterplot, then add a joint density estimate
ax = sns.jointplot(x=iris_df.sepal_length, y=iris_df.sepal_width, data=iris_df)
ax = ax.plot_joint(sns.kdeplot, n_levels=6)
plt.show()
```

```python
# generate random number
x, y = np.random.randn(2, 1000)
```

```python
# pass vector in directly without using pandas then name the axis
g = sns.jointplot(x,y, kind='hex')
g.set_axis_labels("Random X Values", "Random Y Values")
plt.show()
```

```python
# draw the smaller figure with more space devoted to the marginal plot
sns.jointplot("total_bill", "tip", data=tips_df, height=7, ratio=3, color="b")
plt.show()
```

```python
# Stastics Functions importing library
from scipy.stats import spearmanr
```

Calculate a Spearman rank-order correlation coefficient and the p-value to test for non-correlation.

The Spearman correlation is a nonparametric measure of the monotonicity of the relationship between two datasets.

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, stat_func=spearmanr)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='size', y='tip', data=tips_df, stat_func=spearmanr)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, ratio=4, size=6)
plt.show()
```
