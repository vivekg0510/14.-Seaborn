```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.set_style("darkgrid")
```

```python
sns.violinplot(x=tips['tip'])
plt.show()
```

```python
tips.head()
```

```python
sns.violinplot(x=tips["total_bill"])
plt.show()
```

```python
sns.violinplot(x="size", y="total_bill", data=tips)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex')
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex', palette='husl')
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='smoker', palette='husl')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True, palette='coolwarm')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='coolwarm', order=["Sat", "Thur", "Sun", "Fri"])
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='coolwarm', order=["Sat", "Thur", "Sun", "Fri"], hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='RdBu', order=["Sat", "Thur", "Sun", "Fri"], hue='smoker', split=True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='coolwarm', inner='quartile')
plt.show()
```
