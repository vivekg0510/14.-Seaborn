```python
import seaborn as sns, matplotlib.pyplot as plt, numpy as np, pandas as pd
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(3)
```

```python
# the factorplot function has been renamed to catplot. the original name will be removed inn the future release
x = sns.catplot(x='who', y='survived', col='class', data=titanic, kind='bar', ci=None)
x.set_axis_labels('Passenger', 'Survival Rate')
x.set_xticklabels(["Men", "Women", "Children"])
x.set_titles('{col_name}{col_var}')
plt.show()
```
