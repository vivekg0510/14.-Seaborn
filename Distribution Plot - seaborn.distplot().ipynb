# Distribution plot

## seaborn.distplot()

seaborn represents the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bins

this function combine with the matplotlib hist() function

> seaborn.distplot()
> Parameters:-
  1. a :- series, 1D-array, list :- observed data. f this is a series object with a 'name' attribute, the name will be used to label the axis name
  
  2. bins :- argument for matplotlib hist(), None, optional :- specify for hist bins, or none for freedman disconic rule

  3. hist :- bool, optional
  
  4. kde :- bool, optional
  
  5. rug :- bool, optional
  
  6. fit :- random variable object
  
  7. hist_kws :- dictionaries, optional
  
  8. kde_kws  :- dictionaries, optional
  
  9. rug_kws :- dictionaries, optional
  
  10. fit_kws :- dictionaries, optional
  
  11. color :- matplotlib colors, optional
  
  12. vertical :- bool, optional
  
  13. norm_hist :- bool, optional
  
  14. axlabel :- string, False, None, optional
  
  15. label :- String, optional
  
  16. ax :- matplotlib axis, optional

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
```

```python
iris_df = sns.load_dataset("iris")
```

```python
sns.distplot(iris_df['petal_length'], kde=False)
# here kde flag is set to false, as a result the representation of the kde plot will be removed and only histogram is plotted
plt.show()
```

```python
sns.distplot(iris_df['petal_length'], kde=True)
# here kde flag is set to true, as a result the representation of the kde plot and histogram both is plotted
plt.show()
```

```python
#generate a random number 
num = np.random.randn(150)
```

```python
sns.distplot(num)
#plot a histogram using series
```

```python
# set the color on dist plot(histogram plot)
sns.distplot(num, color="red")
```

```python
label_dict = pd.Series(num, name="X Variables")
#set the name of series label
```

```python
sns.distplot(label_dict)
#plot the histogram plot with series and set the x label as 'variable x'
```

```python
sns.distplot(label_dict, vertical=True)
#plot histogram vrtically
plt.show()
```

```python
#plot histogram vertically and set the color as magenta
sns.distplot(label_dict, vertical=True, color="magenta")
```

```python
# plot the distribution plot and remove the histogram
sns.distplot(label_dict, hist=False)
plt.show()
```

```python
# plot distribution plot, support with rugplot and remove histogram plot
sns.distplot(label_dict, rug=True, hist=False)
plt.show()
```

```python
# plot distribution plot, support with rugplot and remove histogram plot and set the color as green
sns.distplot(label_dict, rug=True, hist=False, color="green")
```
