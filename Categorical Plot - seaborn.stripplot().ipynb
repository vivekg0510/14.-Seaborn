```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.stripplot(x=tips['tip'], color='green', jitter=False)
plt.show()
```

```python
sns.stripplot(x=tips['total_bill'], jitter=False)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=0.2)
plt.show()
```

```python
sns.stripplot(x='total_bill', y='day', data=tips, jitter=1)
plt.show()
```

```python
sns.stripplot(x='total_bill', y='day', data=tips, jitter=1, linewidth=2)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, dodge=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True, dodge=True, palette='winter_r', order=['Fri','Sun','Sat','Thur'])
plt.show()
```

```python
sns.stripplot(x='sex', y='tip', data=tips, jitter=False)
plt.show()
```

```python
sns.stripplot(x='sex', y='tip', data=tips, order=['Female','Male'], palette='husl', jitter=False)
plt.show()
```

```python
sns.stripplot(x='sex', y='tip', data=tips, jitter=False, order=['Female', 'Male'], marker='D', palette='husl')
plt.show()
```

```python
sns.stripplot(x='sex', y='tip', data=tips, jitter=False, order=['Female', 'Male'], marker='D', palette='husl', size=15)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, marker='D', size=15, hue='sex', edgecolor='gray', alpha=0.5, jitter=True, dodge=True, palette='husl')
plt.show()
```

```python
sns.stripplot(x='tip', y='day', data=tips, jitter=False)
sns.boxplot(x='tip', y='day', data=tips)
plt.show()
```

```python
sns.stripplot(x='tip', y='day', data=tips, jitter=True, palette='winter_r', color='1')
sns.boxplot(x='tip', y='day', data=tips)
plt.show()
```

```python
sns.stripplot(x='tip', y='day', data=tips, jitter=True)
sns.violinplot(x='tip', y='day', data=tips, color='0.9')
```
