```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(3)
```

```python
sns.boxplot(x=tips['size'])
plt.show()
```

```python
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
sns.boxplot(x='sex', y=tips['total_bill'], data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y=tips['total_bill'], data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y=tips['total_bill'], data=tips, hue='sex', palette='husl')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette='coolwarm')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, order=['Sun','Fri','Thur','Sat'], palette='husl')
plt.show()
```

```python
sns.boxplot(x='sex', y='tip', data=tips, order=['Female','Male'])
plt.show()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.boxplot(data=iris, palette='coolwarm')
plt.show()
```

```python
sns.boxplot(data=iris, orient='horizontal', palette='husl') # horizontal or h
plt.show()
```

```python
sns.boxplot(data=iris, orient='v', palette='husl')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
sns.swarmplot(x='day', y='total_bill', data=tips, color='0.15')
plt.show()
```
