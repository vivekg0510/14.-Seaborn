# Regression Plot

## seaborn.lmplot()

> seaborn.lmplot()

plot data and regression model fits across a FacetGrid. the function combines `regplot()` and `FacetGrid`. 
it is intended as a conventional interface to fit regression model across conditional subset of the dataset

when thinking about how to assign variables to different facets. a general rule is that it makes sese to use `hue` for the most important comparision, followed by `col` and `row`.

> Parameters :-
  1. x,y :- string, optional :- there should be `col` in data.
  2. data :- DataFrame :- Name of df
  3. hue,col,row :- string
  4. palette :- palette name, list, dict, optional
  5. col_wrap :- int, optional
  6. height :-scalar, optional
  7. aspect :- scalar, optional
  8. markers:- matplotlib marker code, optional
  9. sharex, sharey :- bool, col, row, optional
  10. hue_order, col_order, row_order :- lists, optional
  11. legend :- bool, optional
  12. legend_out :- bool, optional
  13. x_estimator :- callable that map vector->scalar, optional 
  14. x_bins :- int, vector, optional
  15. x-ci:- `ci`, `sd`, int in [0,100], None, optional
  16. scatter:- bool, optional
  17. fit_reg:- bool, optional
  18. ci :- int in [0, 100], None, optional
  19. n_boot:- int, optional
  20. units:- variable name or column name of data, optional
  21. order:- int, optional
  22. logistic:- bool, optional
  23. lowess:- bool, optional
  24. robust:- bool, optional
  25. logx:- bool, optional
  26. x_partial, y_partial :- string in data, matrices
  27. truncate :- bool, optional
  28. x_jitter, y_jitter :- floats, optional
  29. scatter_kws, line_kws :- dictionaries

Notes:- the `regplot()` and `lmplot()` function are closely related, but the former is an axes-level function while the latter is a figure-level function that combines `regplot()` and `FacetGrid()`

```python
import seaborn as sns, matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
tips_df = sns.load_dataset("tips")
```

```python
# plot a simple linear relationship between two variable
sns.lmplot(x='total_bill', y='tip',data=tips_df)
plt.show()
```

```python
# conditional on a third variable  and plot the levels in different colors
sns.lmplot(x='total_bill', y='tip',data=tips_df, hue='smoker')
plt.show()
```

```python
# use different marker as well as color so the plot will reproduct to bllack and white more easily
sns.lmplot(x='total_bill', y='tip',data=tips_df, hue='smoker', markers=["o", "x"])
plt.show()
```

```python
#use different color palette
sns.lmplot(x='total_bill', y='tip',data=tips_df, hue='smoker', palette='Set1')
plt.show()
```

```python
#map hue level to colors with a dictionary
sns.lmplot(x='total_bill', y='tip',data=tips_df, hue='smoker', palette=dict(Yes="g", No="m"))
plt.show()
```

```python
#plot the levels of the third variable across different columns
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips_df)
plt.show()
```

```python
#change the height and aspect ratio of the facets
sns.lmplot(x="size", y="total_bill", hue="day", col="day", data=tips_df, height=6, aspect=.4, x_jitter=.1)
plt.show()
```

```python
#wrap the levels of the column variable into multiple rows
sns.lmplot(x="total_bill", y="tip", data=tips_df, col="day",hue="day", col_wrap=2, height=3)
plt.show()
```

```python
#use methods on the return `FacetGrid` instead to further tweak the plot
graph = sns.lmplot(x='total_bill', y="tip", data=tips_df, row="sex", col="time")

graph.set_axis_labels("Total Bills In US Dollers")
plt.show()
```
