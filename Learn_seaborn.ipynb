# figure aesthetic

```python
import numpy as np
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*.5)*(8-i)*flip)
sinplot()
plt.show()
```

```python
# use set() for change the theme
```

```python
sns.set()
sinplot()
plt.show()
```

## seaborn figure style

> five figure style options in seaborn
  1. darkgrid
  2. whitegrid
  3. white
  4. dark

```python
sns.set_style("whitegrid")
sinplot()
plt.show()
```

## remove axes spines

> for removing axes spines use despine() function

```python
sns.set_style("whitegrid")
sinplot()
sns.despine(left=True)
plt.show()
```

# Color palette

> a palette means a flat surface on which a painter arrange and mixed paints

## building color palette

>for building color palette we use the color_palette()

> n_colors :- numbeer of color i the palette. by default the value of n_colors is 6(+4 Updated) colors
> desat :- proporting t desaturate each color

> it return to the list of RGB tuples. some seaborn palette are as follows:-
   1. deep
   2. muted
   3. bright
   4. pastle
   5. dark
   6. colorblind

> it is hard to decide which palette should be used for a given data set without knowing the characterstics of data. seaborn will classify the different ways for using color_palette()

> sns.palplot() which deals with color palettes. the function plots the color palette as horizontal array

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

## sequential color palette

> express the distribution of data ranging from relative lower values to heigher values within a range
> with appending additional character 's'to the color passed to the color parameter will plot the sequential plot.

```python
current_palette = sns.color_palette()
sns.palplot(sns.color_palette("Greens"))
plt.show()
```

## diverging colr palette

> it has two different colors. this is the combination of two variance colors depending on the range point in either direction

> assume plotting the data range from -1 to 1. the alue from -1 to 0 takes one color and 0 to +1 takes another color

```python
sns.palplot(sns.color_palette("BrBG", 9))
plt.show()
```

## setting the default colr palette

> color_palette() has a scompanion called set_palette(). argument are the same for both set_palette() and color_palette()

```python
sns.set_style("whitegrid")
sns.set_palette("husl")
sinplot()
plt.show()
```
