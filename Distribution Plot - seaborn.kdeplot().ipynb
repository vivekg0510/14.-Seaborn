# Distribution Plot

## KDE Plot

> kernal density estimation is a way to estimate the probability density function of a continuous rando variable. it is use for nonn parametric analysis

> setting the hist flag to False in distplot will yield the kernal density estimation plot

> fit and plot a univariate or bivariate kernal density estimate

> parameters in `seaborn.kdeplot()` 
  1. data :- 1D Array :- input data
  2. data2 :- 1D Array, Optional :- second input data
  3. shade :- bool, optional :- if true, shade in the area under the kde curve
  4. vertical :- bool, optional :- if true , density is on x-axis
  5. kernel :- {`gau`, `cos`, `biw`, `epa`, `tri`, `triw`}, optional :- code for shape of kernal to fit with bivariate kde can only use `gau` 
  6. bw :- {`scott`, `silverman`, `scalar`, `pair of scalar`}, optional
  7. gridsize :- int, optional :- number of discrete points 
  8. cut :- scaler, optional :- draw the estimate to `cut*bw`
  9. clip :- pair of scalars, pair of pair scalars, optional :- lower and upper bounds for datapoints use to fit KDE.  
  10. legend :- bool, optional :- if true add a legend or label the axes
  11. cumulative :- bool, optional :- if true , draw the cumulative distribution estimated by the kde
  12. shade_lowest :- bool, optional :- if true shade the lowest contour of a bivariate kde plot , if false when you want multiple densities on the same axis
  13. cbar :- bool, optional :- add a color bar
  14. cbar_ax :- matplotlib axes, optional :- existing axes to daw the colorbar onto , otherwise space is taken from the main axes.
  15. cbar_kws :- dict, optional :- `fig.colorbar()`
  16. ax :- matplotlib axes, optional :- axes to plot on, otherwise uses current axes 
  17. **kwargs :- key, value pairing :- `plt.plot()` or plt.contour{f}

## plotting bivariate distribution

> bivariate distribution is used to determine the relation between two variable. this mainly deals with relationship between two variables and how one variable is behaving with respect to the other.

```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
# set the theme using `set_style()` with `whitegrid` value
```

```python
mean = [0,0]
cov = [[0.2,0],[0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
```

```python
#plot the kdeplot with input a data
sns.kdeplot(x_axis)
plt.show()
```

```python
# plot the kdeplot with input a data with y_axis
sns.kdeplot(y_axis)
plt.show()
```

```python
#plot the kde plot with x_axis value with shade fill with default colors if shade=True
sns.kdeplot(x_axis, shade=True)
plt.show()
```

```python
#plot the kde plot with x_axis value with shade fill with specific colors if shade=True and color="green"
sns.kdeplot(x_axis, shade=True, color='green')
plt.show()
```

```python
#plot the kde plot with x_axis value with shade fill with specific colors if shade=True and color="purple"
sns.kdeplot(y_axis, shade=True, color='purple')
plt.show()
```

```python
#plot the kde plot with x_axis value with shade fill with specific colors if shade=True and color="green"
sns.kdeplot(y_axis, shade=True, color='green')
plt.show()
```

```python
iris_df = sns.load_dataset("iris")
```

```python
iris_df.head(1)
```

```python
setosa = iris_df.loc[iris_df.species=='setosa']
versicolor = iris_df.loc[iris_df.species=='versicolor']
```

```python
# plot the kdeplot with 2 values and set color=green
sns.kdeplot(setosa.petal_length, setosa.petal_width, color='green')
plt.show()
```

```python
# plot the kdeplot with 2 values and set color=green
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, color='green')
plt.show()
```

```python
# plot the kdeplot with 2 values and set color=green in a single figure
sns.kdeplot(setosa.petal_length, setosa.petal_width, color='green')
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, color='green')
plt.show()
```

```python
# plot the kdeplot with 2 values and set cmap=[coolwarm, RdBu] in a single figure
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='coolwarm')
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap='RdBu')
plt.show()
```

```python
# plot the kdeplot with 2 values and set cmap=[Blues_d, RdBu] in a single figure
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='Blues_d')
sns.kdeplot(versicolor.petal_width, versicolor.petal_length, cmap='RdBu')
plt.show()
```

```python
# plot the kdeplot with 2 values and set cmap=[Blues_d, RdBu] in a single figure
#limit the density curve within the range of the data
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='Blues_d', cut=0)
sns.kdeplot(versicolor.petal_width, versicolor.petal_length, cmap='RdBu', cut=0)
plt.show()
```

```python
# plot the kdeplot with 2 values and set cmap=[Blues_d, RdBu] in a single figure
#add a colorbar forr the contours
sns.kdeplot(setosa.petal_width, setosa.petal_length, cmap='Blues_d', cbar=True)
sns.kdeplot(versicolor.petal_width, versicolor.petal_length, cmap='RdBu', cbar=True)
plt.show()
```

```python
ax = sns.kdeplot(setosa.sepal_width,setosa.sepal_length,cmap='Reds',shade=True,shade_lowest=False)
ax = sns.kdeplot(versicolor.sepal_width,versicolor.sepal_length,cmap='Blues',shade=True,shade_lowest=False)
plt.show()
```
