# Distribution Plot

## seaborn.pairplot()

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes.

each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.

drawing a plot to show the univariate distribution of the data for the variable in that column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

This is a high-level interface for PairGrid that is intended to make it easy to draw a few common styles. You should use PairGrid directly if you need more flexibility.

> Parameters:
  1. data :- DataFrame :- Tidy dataframe where each column is a variable and each row is an observation.
  2. hue :- string (variable name), optional :- Variable in data to map plot different colors.
  3. hue_order :- list of strings :- Order for the levels of the hue variable
  4. palette :- dictionary or seaborn color palette :- Set of colors for mapping the hue variable. If a dict, keys should be values in the hue variable.
  5. vars :- list of variable names, optional :- Variables within data to use, otherwise use every column with a numeric datatype.
  6. {x, y}_vars : lists of variable names, optional :- Variables within data to use separately for the rows and columns of the figure; i.e. to make a non-square plot.
  7. kind : {‘scatter’, ‘reg’}, optional :- Kind of plot for the non-identity relationships.
  8. diag_kind : {‘auto’, ‘hist’, ‘kde’}, optional :- Kind of plot for the diagonal subplots. The default depends on whether "hue" is used or not.
  9. markers : single matplotlib marker code or list, optional :- Either the marker to use for all datapoints or a list of markers with a length the same as the number of levels in the hue variable so that differently colored points will also have different scatterplot markers.
  10. height : scalar, optional :- Height (in inches) of each facet.
  11. aspect : scalar, optional :- Aspect * height gives the width (in inches) of each facet.
  12. dropna : boolean, optional :- Drop missing values from the data before plotting.
  13. {plot, diag, grid}_kws : dicts, optional :- Dictionaries of keyword arguments.

Datasets under real-time study contain many variables. In such cases, the relation between each and every variable should be analyzed. Plotting Bivariate Distribution for (n,2) combinations will be a very complex and time taking process.

To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n,2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

We can observe the variations in each plot. The plots are in matrix format where the row name represents x axis and column name represents the y axis.

The diagonal plots are kernel density plots where the other plots are scatter plots as mentioned.

```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the iris dataset
iris_df = sns.load_dataset("iris")
sns.set_style("darkgrid")
```

```python
# plot the pairplot from iris dataset
sns.pairplot(iris_df)
plt.show()

# Draw scatterplots for joint relationships and histograms for univariate 
```
