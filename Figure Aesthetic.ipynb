# Figure Aesthetic

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sinplot(flip = 1):
    x = np.linspace(0,14,100)
    for digit in range(1,5):
        plt.plot(x, np.sin(x+digit*0.5)*(7-digit)*flip)
sinplot()
plt.show()
```

```python
# this is how a plot looks with the default matplotlb
```

```python
# to change the same plot to seaborn default, use the set()
```

```python
import seaborn as sns
```

```python
sns.set()
sinplot()
plt.show()
```

```python
sns.set_style("whitegrid")
sinplot()
plt.show()
```

```python
# removing Axes Spines
```

```python
#in white and ticks theme , we can remove the top and right axes spines using the despine function
```

```python
sinplot()
sns.set_style("white")
sns.despine()
plt.show()
```

```python
# Over riding the elements
```

```python
# if you want to customize the seaborn style, you can pass a dictionary of parameter to the set_style(), parameter viewed using axes_style() function
```

```python
sns.axes_style()
```

```python
sns.set_style("darkgrid", {'axes.axisbelow':False})
sinplot()
sns.despine()
plt.show()
```
